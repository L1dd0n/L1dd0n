```python
import statistics as s
from tabula.io import read_pdf
import pandas as pd
```

Here we have the function that calculates the average and then the on going difference, because if you change one step then the rest change as well so that has been accounted for.

```python
def stair_solution(stairs):
    difference = [] 
    average = round(stairs.mean())
    for i in range(len(stairs)):
        if len(difference) > 0:
            difference.append(((stairs[i]- average) + difference[i-1]))
            # print(difference[i-1])
        else:
            difference.append((stairs[i])- average)
    return difference
```

Now lets read the pdf into a dataframe

```python

def read_stair_pdf():
  pdf_path = "https://github.com/L1dd0n/L1dd0n/raw/main/Risers_in_WM.pdf"
  df_list = read_pdf(pdf_path, pages= "all")
  first_df = pd.concat(df_list[0:], axis =1)
  
  return(first_df)
```

The pdf that my friend sent me was not ready to be intergrated into the stair solution function so i had to clean it up by taking out the unnamed columns, renameing one column as pandas seemed to belive it had the same index as the one prior and then converting the floats to integers. 

```python
def clean_columns(df):
  new_cols = ['454', '455', '456', '457', '458', '459', '460', '461', '461.1',
       '462', '462.1', '463', '463.1', '464', '464.1', '465', '465.1', '466',
       '466.1', '467', '467.1']
  unnamed_cols = df.columns.str.contains('Unnamed')
  df_l = df.drop(df[df.columns[unnamed_cols]], axis=1)
  df_l.columns = new_cols
  df_l.fillna(df_l.mean(), inplace= True)
  df_l = df_l.astype(int)
  return(df_l)
```

Finally i then called the previous functions into a varible which then looped through each column of the new data frame and printed the column name, the average and the differences.

```python
#reading in document and concating all tables 


#calculate stair difference
df_v1 = read_stair_pdf()
df_v2 = clean_columns(df_v1) 
for col in df_v2.columns:
  print("Column name: ", col)
  print("Average: ", round(df_v2[col].mean()))
  print("Differences: ", stair_solution(df_v2[col]))

# ! rm *.*
```
