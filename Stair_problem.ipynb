```python
import statistics as s
from tabula.io import read_pdf
import pandas as pd
```

Now lets read the pdf into a dataframe

```python

def read_stair_pdf():
  pdf_path = "https://github.com/L1dd0n/L1dd0n/raw/main/Risers_in_WM.pdf"
  df_list = read_pdf(pdf_path, pages= "all")
  first_df = pd.concat(df_list[0:], axis =1)
  
  return(first_df)
```

Here we have the function that calculates the average, which will be used as a benchmark to gauge how much each step needs to be changed by, and then the on going difference, because if you change one step then the rest change as well so that has been accounted for. In the beggining I tried to convert the null values to 0's but found that messed with the mean so then I converted the null values to equal the mean but then figured that would give extra values that werent actually there so finally i decided to make the null values 0 then remove them before the stair differences were calculated. Then converted the floats to integers to make the data more readable.

```python
# Write some tests for the results of this code
def stair_solution(stairs):
    difference = [] 
    stairs.dropna(inplace= True)
    stairs = stairs.tolist()
    average = round(s.mean(stairs))
    stairs_v2 = []
    for i in stairs:
      stairs_v2.append(int(i))
    stairs = stairs_v2
    for i in range(len(stairs)):
        if len(difference) > 0:
            difference.append(((stairs[i]- average) + difference[i-1]))
        else:
            difference.append((stairs[i])- average)
    return difference
```

The pdf that my friend sent me was not ready to be intergrated into the stair solution function so i had to clean it up by taking out the unnamed columns, renameing one column as pandas seemed to belive it had the same index as the one prior.

```python
def cleaning_columns(df):
  new_cols = ['454', '455', '456', '457', '458', '459', '460', '461', '461.1',
       '462', '462.1', '463', '463.1', '464', '464.1', '465', '465.1', '466',
       '466.1', '467', '467.1']
  unnamed_cols = df.columns.str.contains('Unnamed')
  df_l = df.drop(df[df.columns[unnamed_cols]], axis=1)
  df_l.columns = new_cols
  # df_l.fillna(0, inplace= True) m
  # df_l = df_l.astype(int)
  return(df_l)
```

Finally i then called the previous functions into a varible which then looped through each column of the new data frame and printed the column name, the average and the differences.

```python
#reading in document and concating all tables 


#calculate stair difference
df_v1 = read_stair_pdf()
df_v2 = cleaning_columns(df_v1) 
for col in df_v2.columns:
  print("Column name: ", col)
  print("Average: ", round(df_v2[col].mean()))
  print("Differences: ", stair_solution(df_v2[col]))

# ! rm *.*
```
